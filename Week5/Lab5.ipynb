```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("titanic_disaster.csv")
```

12. Trực quan thông tin tương quan tỉ lệ sống sót và thiệt mạng trên từng nhóm giới tính

```python
gender_table = pd.crosstab(df["Sex"], df["Survived"])
gender_table
```

```python
plt.figure()
sns.countplot(data=df, x="Sex", hue="Survived")
plt.title("Tỉ lệ sống sót theo giới tính")
plt.xlabel("Giới tính")
plt.ylabel("Số lượng")
plt.show()
```

13. Trực quan thông tin hành khách sống sót trên từng nhóm phân loại hành khách (Pclass).


```python
pclass_table = pd.crosstab(df["Pclass"], df["Survived"])
pclass_table
```

```python
plt.figure()
sns.countplot(data=df, x="Pclass", hue="Survived")
plt.title("Tỉ lệ sống sót theo Pclass")
plt.xlabel("Pclass")
plt.ylabel("Số lượng")
plt.show()
```

14. Trực quan thông tin hành khách sống sót trên từng nhóm giới tính và thang đo tuổi tác

```python
df[["Sex", "Age", "Survived"]].dropna().head(10)
```

```python
plt.figure()
sns.histplot(data=df, x="Age", hue="Survived", bins=30)
plt.title("Phân bố độ tuổi theo trạng thái sống sót")
plt.xlabel("Tuổi")
plt.ylabel("Số lượng")
plt.show()
```

15. Trực quan xác suất hành khách sống sót dựa trên thông tin nhóm đi cùng

```python
df["FamilySize"] = df["SibSp"] + df["Parch"]
df[["FamilySize", "Survived"]].head(10)
```

```python
plt.figure()
sns.barplot(data=df, x="FamilySize", y="Survived")
plt.title("Xác suất sống sót theo quy mô gia đình")
plt.xlabel("Số người đi cùng")
plt.ylabel("Xác suất sống sót")
plt.show()
```

16. Trực quan xác suất hành khách sống sót dựa trên thông tin giá vé


```python
df[["Fare", "Survived"]].describe()
```

```python
plt.figure()
sns.boxplot(data=df, x="Survived", y="Fare")
plt.title("Giá vé theo trạng thái sống sót")
plt.xlabel("Survived (0 = chết, 1 = sống)")
plt.ylabel("Giá vé")
plt.show()
```

17. Trực quan số lượng người thiệt mạng và sống sót theo phân lớp (Pclass) hành khách và
cảng sẽ cập bến.

```python
embark_table = pd.crosstab([df["Pclass"], df["Embarked"]], df["Survived"])
embark_table
```

```python
plt.figure()
sns.countplot(data=df, x="Pclass", hue="Embarked")
plt.title("Phân bố hành khách theo Pclass và cảng lên tàu")
plt.xlabel("Pclass")
plt.ylabel("Số lượng")
plt.show()
```
